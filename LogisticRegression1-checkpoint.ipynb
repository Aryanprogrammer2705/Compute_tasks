```python
def draw_histograms(df, features, rows, cols):
    fig=plt.figure(figsize=(20,20))
    for i, feature in enumerate(features):
        ax=fig.add_subplot(rows,cols,i+1)
        df[feature].hist(bins=20,ax=ax,facecolor='red')
        ax.set_title(feature+"Distribution", color='blue')
    fig.tight_layout()
    plt.show()
draw_histograms(df, df.columns, 6, 3)
```

Columns which have two bars at 0 and 1 have only two unique values(0 and 1) while others which have multiple bars are having guassian distribution or having continuous values. 

```python
sns.countplot(x='TenYearCHD',data=df)
```

No of positive cases are represented by 1 and no of negative cases by 0

```python
X=df.iloc[:,0:14]
Y=df['TenYearCHD']
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=0)
```

```python
lr=LogisticRegression()
lr.fit(X_train,Y_train)
```

```python
X_test_prediction=lr.predict(X_test)
accuracy=accuracy_score(X_test_prediction,Y_test)
accuracy
```

```python
X_train_prediction=lr.predict(X_train)
accuracy=accuracy_score(X_train_prediction,Y_train)
accuracy
```

```python
plot_confusion_matrix(lr,X_test,Y_test)
```

```python
from sklearn.metrics import classification_report
cr=classification_report(Y_test,X_test_prediction)
print(cr)
```

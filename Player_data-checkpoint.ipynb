```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
player_df=pd.read_csv('player_attributes.csv')
player_df
```

```python
name_df=pd.read_csv('player.csv')
name_df
```

```python
player_df=player_df.merge(name_df[['player_name','id','height']])
player_df
```

```python
player_df.drop(['Unnamed: 0','date','player_fifa_api_id','player_api_id'],axis=1,inplace=True)
player_df
```

```python
player_df.isna().sum()
```

```python
sns.heatmap(player_df.isna())
plt.show()
```

```python
player_df.duplicated().sum()
```

```python
rating_analysis=player_df.sort_values('overall_rating',ascending=False)[:10]
rating_analysis=rating_analysis[['overall_rating','player_name']]
plt.figure(figsize=(20,10))
sns.barplot(x='player_name',y='overall_rating',data=rating_analysis)
plt.xlabel('Players',fontsize=16)
plt.ylabel('Rating',fontsize=16)
plt.title('Plot of top 10 rated players',fontsize=18)
plt.legend()
```

```python
potential_analysis=player_df.sort_values('potential',ascending=False)[:10]
potential_analysis=potential_analysis[['potential','player_name']]
plt.figure(figsize=(20,10))
sns.barplot(x='player_name',y='potential',data=potential_analysis)
plt.xlabel('Players',fontsize=16)
plt.ylabel('Potential',fontsize=16)
plt.title('Plot of top 10 potential players',fontsize=18)
plt.legend()
```

```python
penalty_analysis=player_df.sort_values('penalties',ascending=False)[:10]
penalty_analysis=penalty_analysis[['penalties','player_name']]
plt.figure(figsize=(20,10))
sns.barplot(x='player_name',y='penalties',data=penalty_analysis)
plt.xlabel('Players',fontsize=16)
plt.ylabel('Penalties',fontsize=16)
plt.title('Plot of top 10 penalty scoring players',fontsize=18)
plt.legend()
```

```python
plt.figure(figsize=(12,6))
sns.countplot(x='preferred_foot',data=player_df)
plt.xlabel('Preferred Foot',fontsize=16)
plt.ylabel('Count',fontsize=16)
plt.title('Plot of preferred foot among players',fontsize=18)
plt.legend()
```

```python
less_170=player_df[player_df['height']<=170]
less_175=player_df[(player_df['height']>170)&(player_df['height']<=175)]
less_180=player_df[(player_df['height']>175)&(player_df['height']<=180)]
less_185=player_df[(player_df['height']>180)&(player_df['height']<=185)]
less_190=player_df[(player_df['height']>185)&(player_df['height']<=190)]
less_195=player_df[(player_df['height']>190)&(player_df['height']<=195)]
more_195=player_df[(player_df['height']>195)]
```

```python
plt.figure(figsize=(20,10))
data=[less_170['player_name'].count(),less_175['player_name'].count(),less_180['player_name'].count(),less_185['player_name'].count(),less_190['player_name'].count(),less_195['player_name'].count(),more_195['player_name'].count()]
label=['<=170','>170 & <=175','>175 & <=180','>180 & <=185','>185 & <=190','>190 & <=195','>195']
plt.pie(data,labels=label,autopct='%.0f%%')
plt.title('Plot of percentage of players in the height of given range',fontsize=18)
plt.show()
```

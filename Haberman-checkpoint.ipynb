```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('haberman.csv')
```

```python
df
```

```python
df.columns = ['patient_age', 'operation_year', 'positive_axillary_nodes', 'survival_status']
df
```

```python
df.info()
```

```python
df.isna().sum()
```

```python
df['survival_status'].value_counts()
```

```python
df.describe()
```

```python
df['survival_status'].replace([1,2],['Yes','No'],inplace=True)
df
```

```python
sns.countplot(x='survival_status',data=df)
```

```python
operation_year_data=df.groupby('operation_year')[['patient_age']].count().rename(columns={'patient_age':'Count'})
operation_year_data
```

```python
operation_year_data.plot(kind='line')
plt.xlabel('Year',fontsize=18)
plt.ylabel('No. of operations',fontsize=18)
plt.show()
```

```python
operation_survival_data=df.groupby(['operation_year','survival_status'])[['patient_age']].count().rename(columns={'patient_age':'Count'})
operation_survival_data
```

```python
operation_survival_data.reset_index(inplace=True)
operation_survival_data
```

```python
pivot=pd.pivot_table(operation_survival_data[['operation_year','survival_status','Count']],index='operation_year',columns='survival_status',values='Count',aggfunc=np.sum)
pivot.plot(kind='bar',stacked='True',figsize=(15,6))
plt.xlabel('Year',fontsize=16)
plt.title('Distribution of survival status per year',fontsize=18)
```

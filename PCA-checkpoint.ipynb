```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import LabelEncoder
from sklearn.decomposition import PCA
```

```python
df=pd.read_csv('heart.csv')
df
```

```python
df.shape
```

```python
df.describe()
```

```python
df.info()
```

```python
X = df.drop("HeartDisease",axis=1)
y = df.HeartDisease
```

```python
encoder=LabelEncoder()
X['Sex']=encoder.fit_transform(X['Sex'])
X['ChestPainType']=encoder.fit_transform(X['ChestPainType'])
X['RestingECG']=encoder.fit_transform(X['RestingECG'])
X['ExerciseAngina']=encoder.fit_transform(X['ExerciseAngina'])
X['ST_Slope']=encoder.fit_transform(X['ST_Slope'])
X
```

```python
scaler=StandardScaler()
X=scaler.fit_transform(X)
X=pd.DataFrame(X)
X
```

```python
X.columns=['Age','Sex','ChestPainType','RestingBP','Cholesterol','FastingBS','RestingECG','MaxHR','ExerciseAngina','Oldpeak','ST_Slope']
X
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=0)
```

```python
lr=LogisticRegression()
```

```python
lr.fit(X_train,y_train)
```

```python
lr.score(X_test,y_test)
```

```python
lr.score(X_train,y_train)
```

APPLYING PCA

```python
pca=PCA(n_components=5)
```

```python
X2=pca.fit_transform(X)
```

```python
X2.shape
```

```python
pca.components_
```

```python
pca.explained_variance_
```

```python
pca.explained_variance_ratio_
```

```python
X2_train,X2_test,y_train,y_test=train_test_split(X2,y,test_size=0.2,random_state=47)
```

```python
lr=LogisticRegression(max_iter=1000)
```

```python
lr.fit(X2_train,y_train)
```

```python
lr.score(X2_test,y_test)
```

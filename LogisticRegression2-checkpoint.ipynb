```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.metrics import plot_confusion_matrix
```

```python
df=pd.read_csv('Ad Click Data.csv')
df
```

```python
df.info()
```

```python
df.isna().sum()
```

```python
sns.barplot(x='Male',y='Daily Time Spent on Site',data=df)
plt.show()
```

Whether it is male or female,both genders are spending on an average equal time on the site. 

```python
df=df.drop(['Ad Topic Line','City','Country','Timestamp'],axis=1)
df
```

```python
df.isna().sum()
```

```python
df.corr()
```

```python
df['Daily Time Spent on Site']=df.groupby(['Age','Clicked on Ad'])['Daily Time Spent on Site'].apply(lambda x: x.fillna(x.mean()))
df
```

```python
df.isna().sum()
```

```python
df['Area Income']=df.groupby(['Age','Clicked on Ad'])['Area Income'].apply(lambda x: x.fillna(x.mean()))
df
```

```python
df.isna().sum()
```

```python
df['Daily Internet Usage']=df.groupby(['Age','Clicked on Ad'])['Daily Internet Usage'].apply(lambda x: x.fillna(x.mean()))
df
```

```python
df.isna().sum()
```

```python
df=df.drop('Male',axis=1)
df
```

```python
sns.countplot(x='Clicked on Ad',data=df)
plt.show()
```

Target variable data is balanced.

```python
sns.set()
sns.distplot(df['Daily Time Spent on Site'])
plt.show()
```

Most people spend around 75 to 80 mins on the site.

```python
sns.set()
sns.distplot(df['Age'])
plt.show()
```

Most people visiting the site are in the age range of 30 to 35 years.

```python
sns.scatterplot(x=df['Age'],y=df['Daily Time Spent on Site'],hue=df['Clicked on Ad'])
plt.show()
```

The more the time people are spending on the site, the lesser is the possibility to click on the ads with exception seen in the people aged more than 50.

```python
sns.scatterplot(x=df['Age'],y=df['Area Income'],hue=df['Clicked on Ad'])
plt.show()
```

The more is the Area Income lesser is the possibility to click on the ads with exception seen in the people aged more than 50.

```python
sns.scatterplot(x=df['Age'],y=df['Daily Internet Usage'],hue=df['Clicked on Ad'])
plt.show()
```

The more is the Daily Internet Usage lesser is the possibility to click on the ads with exception seen in the people aged more than 50.

```python
X=df.iloc[:,:4]
Y=df['Clicked on Ad']
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.3,random_state=0)
```

```python
lr=LogisticRegression()
```

```python
lr.fit(X_train,Y_train)
```

```python
Y_test_pred=lr.predict(X_test)
accuracy=accuracy_score(Y_test_pred,Y_test)
accuracy
```

```python
Y_train_pred=lr.predict(X_train)
accuracy=accuracy_score(Y_train_pred,Y_train)
accuracy
```

```python
plot_confusion_matrix(lr,X_test,Y_test)
```
